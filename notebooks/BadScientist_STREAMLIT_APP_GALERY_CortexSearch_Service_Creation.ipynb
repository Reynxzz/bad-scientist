````python
import nest_asyncio
import pandas as pd
from llama_index.readers.github import GithubRepositoryReader, GithubClient
import os
from urllib.parse import urlparse

nest_asyncio.apply()

def parse_github_url(url):
    """
    Parse GitHub URL to extract owner and repo
    Example: https://github.com/arnaudmiribel/streamlit-extras
    Returns: ('arnaudmiribel', 'streamlit-extras')
    """
    try:
        url = url.rstrip('/')
        path_parts = urlparse(url).path.strip('/').split('/')
        if len(path_parts) >= 2:
            return path_parts[0], path_parts[1]
        else:
            return None, None
    except Exception as e:
        print(f"Error parsing URL {url}: {str(e)}")
        return None, None

def scrape_github_repo(github_token, owner, repo, branch="main"):
    """
    Scrape a single GitHub repository for .py and .md files
    """
    try:
        github_client = GithubClient(github_token=github_token, verbose=True)

        reader = GithubRepositoryReader(
            github_client=github_client,
            owner=owner,
            repo=repo,
            verbose=False,
            filter_file_extensions=(
                [".py", ".md"],
                GithubRepositoryReader.FilterType.INCLUDE,
            ),
        )

        documents = reader.load_data(branch=branch)

        print(f"Successfully scraped {len(documents)} files from {owner}/{repo}")
        return documents

    except Exception as e:
      try:
        github_client = GithubClient(github_token=github_token, verbose=True)

        reader = GithubRepositoryReader(
            github_client=github_client,
            owner=owner,
            repo=repo,
            verbose=False,
            filter_file_extensions=(
                [".py", ".md"],
                GithubRepositoryReader.FilterType.INCLUDE,
            ),
        )

        documents = reader.load_data(branch="master")

        print(f"Successfully scraped {len(documents)} files from {owner}/{repo}")
        return documents

      except Exception as e:
        print(f"Error scraping {owner}/{repo}: {str(e)}")
        return None

def process_csv_and_scrape(csv_path, github_token):
    """
    Process CSV file containing repository information and scrape each repository
    """
    df = pd.read_csv(csv_path)

    all_documents = []

    for index, row in df.iterrows():
        github_url = row['app_github_link']

        owner, repo = parse_github_url(github_url)

        if not owner or not repo:
            print(f"Skipping invalid GitHub URL: {github_url}")
            continue

        print(f"\nProcessing repository: {owner}/{repo}")

        try:
            documents = scrape_github_repo(github_token, owner, repo)
        except Exception as e:
            print(f"Error scraping {owner}/{repo}: {str(e)}")
            continue

        if documents:
            for doc in documents:
                doc.metadata.update({
                    'app_name': row['app_name'],
                    'app_creator': row['app_creator']
                })
            all_documents.extend(documents)

    return all_documents
````

# Turn Text into Python Code Format

````python
from llama_index.core import Document
from pathlib import Path

def process_content(text, file_extension):
    """
    Process content based on file extension
    """
    if file_extension == '.md':
        return markdown_to_python(text)
    elif file_extension == '.py':
        return text  # Return Python files as-is
    return text

def markdown_to_python(markdown_text):
    """
    Convert markdown to Python with comments, preserving Python code blocks
    """
    lines = markdown_text.split('\n')
    in_python_block = False
    result = []

    # Pattern to detect Python code block starts
    python_block_patterns = [
        '```python',
        '```py',
        '``` python',
        '``` py'
    ]

    for line in lines:
        # Check for Python code block markers
        stripped_line = line.strip()

        # Start of Python block
        if any(stripped_line.startswith(pattern) for pattern in python_block_patterns):
            in_python_block = True
            continue
        # End of Python block
        elif stripped_line == '```' and in_python_block:
            in_python_block = False
            continue

        # Process the line
        if in_python_block:
            # Keep Python code as is
            result.append(line)
        else:
            # Handle markdown content
            if not stripped_line:
                # Empty lines become single #
                result.append('#')
            else:
                # Preserve indentation for markdown text
                leading_spaces = len(line) - len(line.lstrip())
                commented_line = ' ' * leading_spaces + '# ' + line.lstrip()
                result.append(commented_line)

    return '\n'.join(result)

def transform_all_documents(documents):
    """
    Transform a list of Document objects based on their file extensions
    """
    transformed_documents = []

    for doc in documents:
        # Get file extension from metadata
        file_path = doc.metadata.get('file_path', '')
        file_extension = Path(file_path).suffix.lower() if file_path else ''

        # Process the content based on file extension
        transformed_text = process_content(doc.text, file_extension)

        # Create new document with transformed text
        new_doc = Document(
            text=transformed_text,
            metadata=doc.metadata
        )
        transformed_documents.append(new_doc)

    return transformed_documents
````

# Code Chunking Function

````python
import re

def chunk_transformed_text(text: str) -> dict:
    lines = text.split('\n')
    chunks = {}
    chunk_number = 1
    current_chunk = []
    comment_buffer = []
    last_line_was_code = False

    def is_code_block_start(line):
        patterns = [
            r'^\s*import\s+\w+',               # import statement
            r'^\s*from\s+\w+\s+import',        # from import
            r'^\s*def\s+\w+',                  # function definition
            r'^\s*class\s+\w+',                # class definition
            r'^\s*\w+\s*=\s*',                 # assignments
            r'^\s*@\w+',                       # decorators
            r'^\s*if\s+__name__\s*==',         # main block
        ]
        return any(re.match(pattern, line) for pattern in patterns)

    for i, line in enumerate(lines):
        stripped = line.strip()

        # Handle comments
        if stripped.startswith('#'):
            comment_buffer.append(line)
            last_line_was_code = False
            continue

        # Skip empty lines between sections
        if not stripped:
            if last_line_was_code:
                current_chunk.append(line)
            continue

        # If we hit a new code block
        if is_code_block_start(line) and not last_line_was_code:
            if current_chunk:
                chunks[chunk_number] = '\n'.join(current_chunk)
                chunk_number += 1
                current_chunk = []
            current_chunk.extend(comment_buffer)
            current_chunk.append(line)
            comment_buffer = []
            last_line_was_code = True
            continue

        # Continue current code block
        if last_line_was_code:
            current_chunk.append(line)
        # Start a new code block
        else:
            if current_chunk:
                chunks[chunk_number] = '\n'.join(current_chunk)
                chunk_number += 1
            current_chunk = comment_buffer + [line]
            comment_buffer = []
            last_line_was_code = True

    # Add final chunk
    if current_chunk:
        chunks[chunk_number] = '\n'.join(current_chunk)

    return chunks
````

# Chunk Code and Store to Snowflake

````python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core.node_parser import SemanticSplitterNodeParser
import snowflake.connector
from tqdm.auto import tqdm
import numpy as np
import json

def batch_insert_to_snowflake(cursor, snowflake_connector, all_chunks, embeddings, batch_size=1000):
    """
    Batch insert chunks and embeddings into Snowflake
    """
    # Prepare the batch insert query
    insert_query = """
    INSERT INTO streamlit_appgalery_chunks
        (doc_text, file_path, file_name, url, app_name, app_creator, chunk_id, embedding)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
    """

    # Initialize batch data
    batch_data = []
    total_records = len(all_chunks)

    # Process in batches with progress bar
    with tqdm(total=total_records, desc="Storing in Snowflake") as pbar:
        for chunk, embedding in zip(all_chunks, embeddings):
            # Add current record to batch
            batch_data.append((
                chunk['text'],
                chunk['metadata']['file_path'],
                chunk['metadata']['file_name'],
                chunk['metadata']['url'],
                chunk['metadata']['app_name'],
                chunk['metadata']['app_creator'],
                chunk['metadata']['chunk_id'],
                embedding
            ))

            # When batch is full or it's the last batch, execute insert
            if len(batch_data) >= batch_size or (pbar.n + 1) == total_records:
                try:
                    # Execute batch insert
                    cursor.executemany(insert_query, batch_data)
                    snowflake_connector.commit()

                    # Update progress bar
                    pbar.update(len(batch_data))

                    # Clear batch data
                    batch_data = []

                except Exception as e:
                    print(f"Error inserting batch: {str(e)}")
                    # Optionally, implement retry logic here
                    snowflake_connector.rollback()
                    raise e

def chunk_and_store_to_snowflake(documents, connection_params, chunk_function):
    embed_model = HuggingFaceEmbedding("Snowflake/snowflake-arctic-embed-m")

    all_chunks = []
    for doc in tqdm(documents, desc="Chunking documents"):
        chunks = chunk_function(doc.text)
        for chunk_id, chunk_text in chunks.items():
            all_chunks.append({
                'text': chunk_text,
                'metadata': {
                    'file_path': doc.metadata.get('file_path', ''),
                    'file_name': doc.metadata.get('file_name', ''),
                    'url': doc.metadata.get('url', ''),
                    'app_name': doc.metadata.get('app_name', ''),
                    'app_creator': doc.metadata.get('app_creator', ''),
                    'chunk_id': chunk_id
                }
            })

    embeddings = []
    for chunk in tqdm(all_chunks, desc="Generating embeddings"):
        embedding = embed_model.get_text_embedding(chunk['text'])
        embeddings.append(json.dumps(embedding))  # Convert to JSON string

    snowflake_connector = snowflake.connector.connect(**connection_params)
    cursor = snowflake_connector.cursor()

    cursor.execute("""
    CREATE OR REPLACE TABLE streamlit_appgalery_chunks (
        doc_text VARCHAR,
        file_path VARCHAR,
        file_name VARCHAR,
        url VARCHAR,
        app_name VARCHAR,
        app_creator VARCHAR,
        chunk_id INTEGER,
        embedding VARCHAR
    )
    """)

    batch_insert_to_snowflake(
        cursor=cursor,
        snowflake_connector=snowflake_connector,
        all_chunks=all_chunks,
        embeddings=embeddings,
        batch_size=1000
    )

    # for chunk, embedding in tqdm(zip(all_chunks, embeddings), desc="Storing in Snowflake"):
    #     cursor.execute("""
    #     INSERT INTO streamlit_appgalery_chunks (doc_text, file_path, file_name, url, app_name, app_creator, chunk_id, embedding)
    #     VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
    #     """, (
    #         chunk['text'],
    #         chunk['metadata']['file_path'],
    #         chunk['metadata']['file_name'],
    #         chunk['metadata']['url'],
    #         chunk['metadata']['app_name'],
    #         chunk['metadata']['app_creator'],
    #         chunk['metadata']['chunk_id'],
    #         embedding
    #     ))
    #     snowflake_connector.commit()

    snowflake_connector.close()
    return all_chunks, embeddings
````

````python
cursor.execute(f"""
CREATE OR REPLACE CORTEX SEARCH SERVICE {cortex_search_service_name}
ON doc_text
WAREHOUSE = COMPUTE_WH
TARGET_LAG = '1 hour'
AS
    SELECT
        doc_text,
        app_name,
        app_creator,
        url,
    FROM {snowflake_table_name}
""")
````

````python
import os
from snowflake.core import Root
from typing import List

class CortexSearchRetriever:

    def __init__(self, session: Session, limit_to_retrieve: int = 4):
        self._session = session
        self._limit_to_retrieve = limit_to_retrieve

    def retrieve(self, query: str) -> List[str]:
        root = Root(self._session)
        cortex_search_service = (
        root
        .databases['BADSCIENTIST_DEMO']
        .schemas['NOTEBOOKS']
        .cortex_search_services[f'{cortex_search_service_name}']
    )
        resp = cortex_search_service.search(
                query=query,
                columns=["doc_text", "app_name", "app_creator", "url"],
                limit=self._limit_to_retrieve,
            )

        if resp.results:
            return [
                {
                    "doc_text": curr["doc_text"],
                    "app_name": curr["app_name"],
                    "app_creator": curr["app_creator"],
                    "url": curr["url"]
                }
                for curr in resp.results
            ]
        else:
            return []
````

## RAG APP

````python
import os
import snowflake.connector
from snowflake.core import Root
from snowflake.snowpark.context import get_active_session

# Constants
DB = "BADSCIENTIST_DEMO"
SCHEMA = "NOTEBOOKS"
SERVICE = f"{cortex_search_service_name}"
BASE_TABLE = f"BADSCIENTIST_DEMO.NOTEBOOKS.{snowflake_table_name}"

# Model options
MODELS = [
    "mistral-large2",
    "snowflake-arctic",
    "llama3-70b",
    "llama3-8b",
]
````

````python
class SnowflakeRagApp:
    def __init__(self, connection_params, model_name="mistral-large2", num_chunks=5):
        """Initialize the RAG application with Snowflake connection and parameters."""
        self.session = get_active_session()
        self.root = Root(self.session)
        self.model_name = model_name
        self.num_chunks = num_chunks
        self.search_column = self.get_service_search_column(SERVICE)

    def get_service_search_column(self, service_name):
        """Get search column for specific service."""
        try:
            desc_result = self.session.sql(
                f"DESC CORTEX SEARCH SERVICE {service_name};"
            ).collect()
            return desc_result[0]["search_column"]
        except Exception as e:
            raise ValueError(f"Error getting search column for service {service_name}: {str(e)}")

    def query_cortex_search(self, query, service_name=SERVICE):
        """Query the cortex search service and return context."""
        db, schema = self.session.get_current_database(), self.session.get_current_schema()

        cortex_search_service = (
            self.root.databases[db]
            .schemas[schema]
            .cortex_search_services[service_name]
        )

        context_documents = cortex_search_service.search(
            query, columns=[], limit=self.num_chunks
        )
        results = context_documents.results

        context_str = ""
        for i, r in enumerate(results):
            context_str += f"Context document {i+1}: {r[self.search_column]} \n\n"

        return context_str

    def complete(self, prompt):
        """Generate completion using Snowflake's language model."""
        return self.session.sql(
            "SELECT snowflake.cortex.complete(?,?)",
            (self.model_name, prompt)
        ).collect()[0][0]

    def create_prompt(self, user_question, context):
        """Create a prompt for the language model."""
        return f"""
        [INST]
        You are a helpful AI chat assistant with RAG capabilities. Use the context provided
        between <context> and </context> tags to provide a summary that addresses the user's
        question. Ensure the answer is coherent, concise, and directly relevant to the
        user's question.

        If the user asks a generic question which cannot be answered with the given context,
        just say "I don't know the answer to that question."

        Don't say things like "according to the provided context".

        <context>
        {context}
        </context>
        <question>
        {user_question}
        </question>
        [/INST]
        Answer:
        """

    def ask(self, question):
        """Process a user question and return the response."""
        # Get context from cortex search
        context = self.query_cortex_search(question)

        # Create prompt with context
        prompt = self.create_prompt(question, context)

        # Generate response
        response = self.complete(prompt)

        return {
            'question': question,
            'context': context,
            'response': response
        }
````
