```python
import nest_asyncio
nest_asyncio.apply()

from llama_index.readers.github import GithubRepositoryReader, GithubClient

github_token = input("Enter GITHUB_TOKEN: ")
github_client = GithubClient(github_token=github_token, verbose=False)

reader = GithubRepositoryReader(
  github_client=github_client,
  owner="streamlit",
  repo="docs",
  use_parser=False,
  verbose=True,
  filter_directories=(
    ["content"],
    GithubRepositoryReader.FilterType.INCLUDE,
  ),
  filter_file_extensions=(
    [".md"],
    GithubRepositoryReader.FilterType.INCLUDE,
  )
)

documents = reader.load_data(branch="main")
```

```python
import re

def clean_up_text(content: str) -> str:
  """
  Remove unwanted characters and patterns in text input.

  :param content: Text input.

  :return: Cleaned version of original text input.
  """

  # Fix hyphenated words broken by newline
  content = re.sub(r'(\w+)-\n(\w+)', r'\1\2', content)

  unwanted_patterns = ['---\nvisible: false','---', '#','slug:']
  for pattern in unwanted_patterns:
    content = re.sub(pattern, "", content)

  # Remove all slugs starting with a \ and stopping at the first space
  content = re.sub(r'\\slug: [^\s]*', '', content)

  # normalize whitespace
  content = re.sub(r'\s+', ' ', content)
  return content
```

```python
from llama_index.core import Document

cleaned_documents = []
for d in documents:
    cleaned_text = clean_up_text(d.text)
    new_doc = Document(
        text=cleaned_text,
        metadata=d.metadata
    )
    cleaned_documents.append(new_doc)
```

### CortexSearchRetriever

```python
import os
from snowflake.core import Root
from typing import List

class CortexSearchRetriever:

    def __init__(self, session: Session, limit_to_retrieve: int = 4):
        self._session = session
        self._limit_to_retrieve = limit_to_retrieve

    def retrieve(self, query: str) -> List[str]:
        root = Root(self._session)
        cortex_search_service = (
        root
        .databases['BADSCIENTIST_DEMO']
        .schemas['NOTEBOOKS']
        .cortex_search_services['streamlit_docs_svc']
    )
        resp = cortex_search_service.search(
                query=query,
                columns=["doc_text"],
                limit=self._limit_to_retrieve,
            )

        if resp.results:
            return [curr["doc_text"] for curr in resp.results]
        else:
            return []
```

## RAG APP

```python
import os
import snowflake.connector
from snowflake.core import Root
from snowflake.snowpark.context import get_active_session

# Constants
DB = "BADSCIENTIST_DEMO"
SCHEMA = "NOTEBOOKS"
SERVICE = "STREAMLIT_DOCS_SVC"
BASE_TABLE = "BADSCIENTIST_DEMO.NOTEBOOKS.STREAMLIT_DOCS"

# Model options
MODELS = [
    "mistral-large",
    "snowflake-arctic",
    "llama3-70b",
    "llama3-8b",
]

class SnowflakeRagApp:
    def __init__(self, connection_params, model_name="mistral-large", num_chunks=5):
        """Initialize the RAG application with Snowflake connection and parameters."""
        self.session = get_active_session()
        self.root = Root(self.session)
        self.model_name = model_name
        self.num_chunks = num_chunks
        self.init_service_metadata()

    def init_service_metadata(self):
        """Initialize cortex search service metadata."""
        services = self.session.sql("SHOW CORTEX SEARCH SERVICES;").collect()
        self.service_metadata = []
        if services:
            for s in services:
                svc_name = s["name"]
                svc_search_col = self.session.sql(
                    f"DESC CORTEX SEARCH SERVICE {svc_name};"
                ).collect()[0]["search_column"]
                self.service_metadata.append(
                    {"name": svc_name, "search_column": svc_search_col}
                )

    def query_cortex_search(self, query, service_name=SERVICE):
        """Query the cortex search service and return context."""
        db, schema = self.session.get_current_database(), self.session.get_current_schema()

        cortex_search_service = (
            self.root.databases[db]
            .schemas[schema]
            .cortex_search_services[service_name]
        )

        context_documents = cortex_search_service.search(
            query, columns=[], limit=self.num_chunks
        )
        results = context_documents.results

        search_col = [s["search_column"] for s in self.service_metadata
                     if s["name"] == service_name][0]

        context_str = ""
        for i, r in enumerate(results):
            context_str += f"Context document {i+1}: {r[search_col]} \n\n"

        return context_str

    def complete(self, prompt):
        """Generate completion using Snowflake's language model."""
        return self.session.sql(
            "SELECT snowflake.cortex.complete(?,?)",
            (self.model_name, prompt)
        ).collect()[0][0]

    def create_prompt(self, user_question, context):
        """Create a prompt for the language model."""
        return f"""
        [INST]
        You are a helpful AI chat assistant with RAG capabilities. Use the context provided
        between <context> and </context> tags to provide a summary that addresses the user's
        question. Ensure the answer is coherent, concise, and directly relevant to the
        user's question.

        If the user asks a generic question which cannot be answered with the given context,
        just say "I don't know the answer to that question."

        Don't say things like "according to the provided context".

        <context>
        {context}
        </context>
        <question>
        {user_question}
        </question>
        [/INST]
        Answer:
        """

    def ask(self, question):
        """Process a user question and return the response."""
        # Get context from cortex search
        context = self.query_cortex_search(question)

        # Create prompt with context
        prompt = self.create_prompt(question, context)

        # Generate response
        response = self.complete(prompt)

        return {
            'question': question,
            'context': context,
            'response': response
        }
```

```python
# Initialize the app
app = SnowflakeRagApp(connection_params)

# Ask individual questions
result = app.ask("How do I code in streamlit?")
print(result['context'])
print(result['response'])
```
